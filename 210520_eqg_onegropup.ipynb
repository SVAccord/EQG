```python
def listToString(list1):
    return (''.join(map(str, list1)))
def listToString_cmm(list1):
    return (','+','.join(map(str, list1))+',')

def list_getsome(vec, positions):
    list_n = []
    for y in positions:
        list_n.append(vec[y])
    #print("lnf", list_n)
    return(list_n)

# convert into different bases
dim = 4
np.base_repr(11111100, base=dim)
int('3333333333', base=dim)
int('33333333', base=dim)

def diversity_num(vec, dim):
    dvst = 0
    for i in range(0, dim):
        for j in range(0, len(vec)):
            if vec[j] == i:
                dvst += 1
                break
    return dvst
vec1 = [np.nan, 1,np.nan, np.nan, 1, 2, np.nan, np.nan, np.nan, np.nan, 2, np.nan]
print(diversity_num(vec1, 3))
def diversity_vec(vec, dim):
    dvst = []
    for i in range(0, dim):
        for j in range(0, len(vec)):
            if vec[j] == i:
                dvst.append(i)
                break
    return dvst
print(diversity_vec(vec1, 3))
```
